<h1> <font color="yellow"> 3.A : undersampling(deleting) means reducing the Major class samples to match minor class sampels <br>
by subtracting or deleting data <br>
<br>

# ***It can be done by deleting some of the sample from major class ,to balance with the minor class*** <br>
for example, major class = 300 , minor class = 200 <br>
so after undersampling we get, major class =200 , minor class = 200 <br>
<br>
question : How do we know which is majority class and which is minority class? <br>
ans : we can look it up, who's got the highest number is manjority afte we check :
  </font> </h1>


```python
from sklearn.utils import resample

majority_class = df[df['class'] == 1 ]
minority_class = df[df['class'] == 0 ]

n_samples = len(minority_class)

majority_downsampled = resample(majority_class ,replace=False,n_samples=n_samples,random_state=42)

balanced_df = pd.concat([minority_class,majority_downsampled])
```

<h1> <font color="yellow">  #now we are printing to see those majority and minority are now same or not   </font> </h1>

```python
#same sortcut code as before to find unique value

value_counts = balanced_df.groupby('class').size().reset_index(name='Count')
print(value_counts)
```
