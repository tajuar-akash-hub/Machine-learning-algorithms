```python
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd

# Prepare the data
data = {
    'Height': [7, 5, 7.1, 7.2, 5.1, 5.2, 4.8],
    'Weight': [40, 50, 42, 42, 52, 55, 50],
    'Class': ['elf', 'dwarf', 'elf', 'elf', 'dwarf', 'dwarf', 'dwarf']
}

df = pd.DataFrame(data)

# Encode target labels
df['Class'] = df['Class'].map({'elf': 0, 'dwarf': 1})  # 0 = elf, 1 = dwarf

# Features and target
X = df[['Height', 'Weight']]
y = df['Class']

# Train decision tree
clf = tree.DecisionTreeClassifier(criterion='gini', max_depth=None)
clf = clf.fit(X, y)


```

```python
import numpy as np

# New data: list of [Height, Weight]
new_data = np.array([
    [6.5, 45],   # likely Elf
    [5.0, 60],   # likely Dwarf
    [7.3, 41],   # likely Elf
    [4.9, 49]    # likely Dwarf
])

# Predict
predictions = clf.predict(new_data)

# Map predictions back to labels
# label_map = {0: 'elf', 1: 'dwarf'}
# predicted_labels = [label_map[p] for p in predictions]
# print(predicted_labels)

print(predictions)
```

```python
# Plot the tree
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=['Height', 'Weight'], class_names=['Elf', 'Dwarf'], filled=True)
plt.title("Decision Tree Based on Height and Weight")
plt.show()
```
